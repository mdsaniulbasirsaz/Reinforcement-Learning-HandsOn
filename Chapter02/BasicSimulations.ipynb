# Basic Simulations in OpenAI's Gym
OpenAI Gym is a toolkit for building, evaluating and comparing RL algorithms. It is compatible with algorithms written in any frameworks like TensoFlow, Theano, Keras etc... It is simple and easy to comprehend. It makes no assumption about the structure of our agent and provides an interface to all RL tasks.

```python
import gymnasium as gym
```

```python
env = gym.make("CartPole-v1")
```

```python
env.reset()
```

```python
for _ in range(1000):
    env.render()
    action = env.action_space.sample() 
    env.step(action)  
```

```python
env = gym.make("CartPole-v1", render_mode="human")
state, info = env.reset()
```

```python
for _ in range(1000):
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        state, info = env.reset()


env.close()
```
