```python
import tensorflow as tf
```

```python
a = tf.constant(1.0)
b = tf.constant(2.0)
c = tf.add(a, b)
print(c.numpy())
```

```python
d = tf.multiply(a, b)
print(d.numpy())
```

```python
with tf.name_scope("Computation"):
    a = tf.constant(5)
    b = tf.constant(4)
    c = tf.multiply(a,b)
    d = tf.constant(2)
    e = tf.constant(3)
    f = tf.multiply(d,e)
```

```python
with tf.name_scope("Result"):
     g = tf.add(c,f)
```

```python
with tf.name_scope("Computation"):
    with tf.name_scope("Part1"):
        a = tf.constant(5)
        b = tf.constant(4)
        c = tf.multiply(a,b)
    with tf.name_scope("Part2"):
        d = tf.constant(2)
        e = tf.constant(3)
        f = tf.multiply(d,e)
```
